```python
import torch
import pathlib
from transformers import MarianMTModel, MarianTokenizer
```

```python
path = pathlib.Path().resolve().parent

tokenizer = MarianTokenizer.from_pretrained(path)
model = MarianMTModel.from_pretrained(path)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

def translate(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    out = model.generate(**inputs, num_beams=5, max_length=128)
    return tokenizer.decode(out[0], skip_special_tokens=True)
```

```python
translate("hello world, this is a machine learning model")
```

```python
translate("I usually eat bread and butter before leaving for work")
```

```python
translate("I hope I get better cards in my game.")
```
